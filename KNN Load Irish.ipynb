# Train test split

```python
from sklearn.model_selection import train_test_split
```

```python
x = df.drop(["target","flower_name"],axis = "columns")
y = df.target
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42) # important for interview point of view.
```

```python
# 42 IS THE DATA USED SO THAT BIAS NA HO MEANS EK JAGAH KE DATA SE DECIDE NA KIA JAYE.
```

```python
len(x_train)
```

#Create KNN

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
KNeighborsClassifier(n_neighbors=5)
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
```

```python
knn.fit(x_train,y_train)
```

```python
knn.score(x_test,y_test)
```

```python
# 1.0 shows 100% accuracy of examining flowers as it shows 150 bar pehchaan chuke hain flowers ko so ab jo bhi new flower hoga usko hm 100% shi hi btayenge accuracy k sath ki o knsa category ka flower hai.
```

```python
from sklearn.metrics import confusion_matrix
y_pred = knn.predict(x_test)
cm = confusion_matrix(y_test,y_pred)
```

```python
cm
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as nk
plt.figure(figsize=(10,5))
nk.heatmap(cm,annot=True) # annot ki wajah se values aati hain.
plt.xlabel("Predicted")
plt.ylabel("Truth")
```
